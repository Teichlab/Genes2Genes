# Supplementary Notebook 1: Checking total time of alignment

```python
import anndata
import numpy as np
import seaborn as sb
import numpy as np
import platform
import warnings
import time
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")

from genes2genes import Main
from genes2genes import VisualUtils
from genes2genes import ClusterUtils
from genes2genes import TimeSeriesPreprocessor
from genes2genes import PathwayAnalyser
from genes2genes import VisualUtils
```

```python
input_dir = 'data/'
adata_ref = anndata.read_h5ad(input_dir + 'adata_pam_local.h5ad') # Reference dataset
adata_query = anndata.read_h5ad(input_dir +'adata_lps_local.h5ad') # Query dataset
# define the gene list to align
gene_list = adata_ref.var_names 
print(len(gene_list),'genes')
```

### A simple experiment to check the number of interpolation time points vs. approximate time taken for alignment

```python
times = []
for n_bins in range(5,50):
    s = time.time()
    aligner = Main.RefQueryAligner(adata_ref, adata_query, gene_list, n_bins)
    aligner.align_all_pairs() 
    t = time.time()
    times.append(t-s) 
```

```python
sb.lineplot(x=range(5,50), y=times, linewidth=3, marker = 'o')
plt.xlabel('Number of interpolation points')
plt.ylabel('Time (seconds)')
plt.title('PAM vs LPS alignment (179 reference cells & 290 query cells in terms of 89 genes)')
#plt.savefig('n_interpolation_points_vs_time_PAM_LPS_G2G_alignment.png')
```

### A simple experiment to check the ref and query dataset size (number of cells) vs the approximate time taken for alignment

```python
adata_ref_mult = [adata_ref]

temp = adata_ref
for i in range(0,10):
    temp = anndata.concat([temp, temp])
    adata_ref_mult.append(temp)
adata_query_mult = [adata_query]
temp = adata_query
for i in range(0,10):
    temp = anndata.concat([temp, temp])
    adata_query_mult.append(temp)

r_size = []
q_size = []
for a in  adata_ref_mult:
    r_size.append(a.shape[0])
for a in  adata_query_mult:
    q_size.append(a.shape[0])

times = []
n_bins = 14
for i in range(len(adata_ref_mult)):
    R = adata_ref_mult[i]
    Q = adata_query_mult[i]
    s = time.time()
    aligner = Main.RefQueryAligner(R, Q, gene_list, n_bins)
    aligner.align_all_pairs() 
    t = time.time()
    times.append(t-s) 

plt.subplots(1,2, figsize=(10,3))
plt.subplot(1,2,1)
sb.lineplot(x=range(len(adata_ref_mult)), y=times, linewidth=3, marker = 'o')
plt.xlabel('k (for aligning genes of $N_{R}.2^k$ cells and $N_{Q}.2^k$ cells)')
plt.ylabel('Time (seconds)')
plt.xticks(range(0,11))
plt.title('Total alignment time of 89 genes')
plt.subplot(1,2,2)
sb.lineplot(x=r_size, y=q_size, linewidth=3, marker = 'o', color='orange')
plt.title('# of cells')
plt.xlabel('Number of reference cells')
plt.ylabel('Number of query cells')
plt.xticks(rotation =10)
plt.tight_layout() 
plt.savefig('cell_numbers_vs_approx_time_PAM_LPS_G2G_alignment.png')
plt.show()
```
